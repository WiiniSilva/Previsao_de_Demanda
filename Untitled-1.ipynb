As lojas enfrentam diversos desafios quando se trata de previsão de demanda, e a falta de previsões precisas pode ter consequências significativas para o desempenho comercial. A previsão de demanda mal feita ou inexistente pode levar a uma série de problemas que impactam tanto a eficiência operacional quanto a satisfação do cliente.

* **Estoque excessivo ou escasso**: Uma previsão de demanda imprecisa pode resultar em excesso de estoque, onde produtos que não estão vendendo conforme esperado ocupam espaço valioso e podem levar a custos adicionais de armazenamento. Por outro lado, a falta de produtos em estoque pode resultar em vendas perdidas, já que os clientes, ao não encontrarem o que procuram, podem optar por comprar em concorrentes. Isso não apenas afeta a receita imediata, mas também pode prejudicar a lealdade do cliente a longo prazo.

* **Custos operacionais elevados**: A falta de previsões adequadas pode levar a um planejamento inadequado da força de trabalho. Se uma loja não consegue prever com precisão os períodos de alta demanda, pode acabar subestimando o número de funcionários necessários, resultando em longas filas e atendimento insatisfatório. Alternativamente, a superestimação de demanda pode resultar em custos desnecessários com mão de obra.

* **Perda de oportunidades de vendas**: Quando as lojas não conseguem antecipar as tendências de consumo, perdem oportunidades de lançar promoções ou campanhas de marketing em momentos críticos. Isso pode resultar em um impacto negativo nas vendas, especialmente em períodos sazonais, como datas comemorativas, quando a demanda costuma ser maior.

* **Dificuldades em identificar tendências de mercado**: A falta de análise de dados históricos para embasar as previsões pode levar as lojas a perderem de vista tendências emergentes no comportamento do consumidor. Isso é especialmente problemático em mercados dinâmicos, onde as preferências dos clientes podem mudar rapidamente. Se uma loja não está ciente dessas mudanças, pode continuar a investir em produtos que não estão mais em demanda.

* **Impacto nas relações com fornecedores**: Previsões de demanda imprecisas podem afetar a relação com fornecedores, especialmente se pedidos excessivos ou insuficientes forem realizados. Isso pode gerar rupturas na cadeia de suprimentos, prejudicando a capacidade da loja de atender à demanda dos clientes de forma eficaz.

* **Efeitos na estratégia de marketing**: Campanhas de marketing mal direcionadas, baseadas em previsões de demanda errôneas, podem levar a investimentos financeiros ineficazes. As lojas podem gastar recursos em publicidade para produtos que não estão em alta, resultando em retorno sobre investimento (ROI) insatisfatório.

Em um ambiente de varejo competitivo, onde a experiência do cliente e a eficiência operacional são fundamentais, a capacidade de realizar previsões de demanda precisas torna-se uma habilidade crítica. As lojas que negligenciam esse aspecto correm o risco de enfrentar sérios desafios, comprometendo sua posição no mercado e sua capacidade de se adaptar às necessidades dos consumidores. Portanto, investir em sistemas de análise de dados e técnicas de previsão se torna não apenas uma vantagem competitiva, mas uma necessidade para a sustentabilidade a longo prazo.

# 5 - Estacionariedade

Existem diversos testes para verificar se uma série temporal é estacionária ou não. Para o nosso problema iremos utilizar os dois mais comuns, que são, o teste de Dickey-Fuller aumentado (ADF) e o teste KPSS (Kwiatkowski-Philips-Schmidt-Shin).

Teste de Dickey-Fuller aumentado:

É um teste estatistico utilizado para veificar a presença de raiz unitária em uma série temporal. uma série temporal é considerada ter uma raiz unitária se ela não for estacionária, ou seja, se sua média e variância não forem constantes ao longo do tempo. essa verificação acontece comparando a série temporal original com uma versão diferenciada dela mesma.

Em resumo, o teste nos ajuda a entender se uma série temporal é “bagunçada” (com raiz unitária) ou “comportada” (estacionária). Se a série for estacionária, podemos fazer previsões mais confiáveis com base nela.

* H0(p-valor > 0.05): A série não é estacionária.
* H1(p-valor <= 0.05): A série é estacionária.

Teste de KPSS:

Compara a série temporal com um modelo estacionário que tem caracteristicas semelhantes à série. Ele avalia a diferença entre a série real e o modelo estacionário para determinar se a série é estacionária ou não.
* H0(p-valor >= 0.05): A série é estacionária.
* H1(p-valor < 0.05): A série não é estacionária.

### 5.1.3 - Conclusões

    * O teste de Dickey-Fuller é um teste de hipótese onde a hipótese nula é que a série não é estacionária. Se o teste rejeita a hipótese nula, sugere que a série é estacionária.

    * O teste KPSS é o oposto: a hipótese nula é que a série é estacionária. Se o teste rejeita a hipótese nula, sugere que a série não é estacionária.

Portanto, no caso em que o teste de Dickey-Fuller rejeita a não estacionariedade apenas em níveis de significância menos rigorosos, e o KPSS rejeita a estacionariedade em todos os níveis de significância, a conclusão mais forte é que a série provavelmente não é estacionária, possivelmente contendo uma tendência ou sazonalidade que deve ser ajustada.

# 6 - Transformando a série em estacionária

Para tornar a série estacionária, iremos fazer uma diferenciação que é uma técnica comum utilizada na análise de séries temporais. Sendo importante porque muitos modelos de previsão, como ARIMA e SARIMA, exigem que os dados sejam estacionários para fornecer previsões precisas.

Mas Por Que Diferenciar?

A série que está sendo analisada foi identificada como não estacionária, tanto pelos testes Dickey-Fuller quanto KPSS. A não estacionariedade significa que as propriedades estatísticas da série (média, variância) mudam ao longo do tempo, o que pode dificultar a modelagem e a previsão. Aplicando a diferenciação, podemos transformar a série em estacionária, removendo essas tendências.

E Como Podemos Fazer a Diferenciação?

Podemos aplicar a diferenciação de várias formas, dependendo da complexidade da série. A diferenciação simples calcula a diferença entre valores consecutivos. Se a série ainda não for estacionária, pode ser necessário aplicar a diferenciação mais de uma vez.

## 7.1 Determinando o valor de AR(p), MA(q), SAR(P) e SMA(Q)

os valores p, q, P e Q são parâmetros fundamentais que ajudam a definir a estrutura do modelo.

* **p**: representa a ordem do componente auto-regressivo (AR) do modelo. Isso significa que o valor p indica quantos termos passados de lag da série são usados para prever os valores futuros. Se definirmos p = 2, o modelo usará os dois períodos de tempo anteriores para prever o próximo valor. Esse componente é crucial para capturar a relação entre os valores da série ao longo do tempo.

* **q**: representa a ordem do componente de média móvel (MA) do modelo. Isso significa que o valor q
indica quantos termos passados do erro são usados para prever os valores futuros. Definindo o q = 1, o modelo usará o erro do período de tempo anterior para ajustar a previsão atual. Esse componente é essencial para corrigir erros de previsão e capturar a influência dos choques passados.

* **P**: indica quantos períodos sazonais passados (lags sazonais) serão usados para prever os valores futuros. Por exemplo, se a sazonalidade é anual e P = 1, o modelo usará os dados da mesma estação do ano anterior para prever os valores futuros.

* **Q**: representa a ordem do componente de média móvel sazonal no modelo SARIMA. Ele é fundamental para capturar o impacto dos erros passados em períodos sazonais anteriores sobre a previsão atual.

Para determinarmos o valor de p e P da autoregressão utilizaremos a PACF:

Box e Jenkins criaram alguns critérios que funcionam em boa parte dos casos, critérios representados na tabela a seguir:

| ACF | PACF | Modelo a ser tentado |
| --- | --- | --- |
| Queda gradual ou sinusal até zero | Queda abrupta após lag p | AR(p) |
| Queda abrupta após lag q | Queda gradual ou sinusal até zero | MA(q) |
| Queda abrupta após lag q | Queda abrupta após lag p | Escolha o melhor entre AR(p) e MA(q) |
| Queda gradual ou sinusoidal até zero | Queda gradual ou sinusoidal até zero | Modelo ARMA |
| Sem picos expressivos | Sem picos expressivos | Série é randômica ou ruído branco |
| Picos em intervalos fixos | Picos em intervalos fixos | A série possui sazionalidade |
| Não decai até zero ou decai muitos lags | | Série não é estacionária |
